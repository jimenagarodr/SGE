# Matplotlib

Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1) 
y = x*x
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker = '*', linestyle = 'dotted',linewidth = '2.5')
# linewidth es el ancho
plt.show()
```

```python
plt.plot(x,y,marker = '*', linestyle = ':',linewidth = '2.5',color='g',label='$x^2$')
plt.plot(x,x*8,marker = '+', linestyle = '--',linewidth = '1.5',color='r',label='8*x')
font = {'family':'serif','color':'darkred','size':15}
plt.title("Título", loc = 'center', fontdict = font)
plt.xlabel("x")
plt.ylabel("y = $x^2$")
plt.legend()
plt.grid()
plt.show()
```

### Scatter

Para el siguiente gráfico "scatter" utizaremos un cmap, los cmap disponibles están en la siguiente página:  
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
#gráfico de burbujas o grafico de bolas
def f(i,j):
    return (j**2)*(i**2)

x=[]
y=[]
z=[]
N=10
step=2
for i in np.arange(-N, N+step, step):
    for j in np.arange(-N, N+step, step):
        x.append(i)
        y.append(j)
        z.append(f(i,j))
        
plt.scatter(x, y, c=z, cmap='viridis',s=z, alpha=0.5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
plt.show()
```

### Bars and Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Soleado", "Nublado", "Lluvia", "Nieve"])
y = np.array([70, 20, 9, 1])

plt.bar(x,y)
plt.show()
```

```python
plt.barh(x, y, height = 0.1,color = "red")
plt.show()
```

```python
plt.pie(y, labels = x)
plt.show()
```

```python
plt.pie(y, labels = x,startangle = 90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4,5,0.1) 
y = (12/(x*x + 1))
plt.plot(x,y)

plt.plot(x,y,color='g',label=r'$\frac{12}{(x*x+1)}$')

font = {'family':'serif','color':'blue','size':16}
plt.title("Jimena García :)", loc = 'center', fontdict = font)

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
from math import e
x = np.arange(-10,10, 0.1) 
plt.plot(x,(1/(1+e**-x)),color='g',label=r'$\frac{1}{1+e^{-x}}$')

plt.axvline(-2.5,ymax= 0.95, ymin=0.05, color = "blue", linestyle = "dashed");
plt.axvline(x=2.5,ymax= 0.95, ymin=0.05, color = "blue", linestyle = "dashed");
plt.axhline(y=0, xmin=0.39, xmax=0.62, c="blue", linestyle = "dashed");
plt.axhline(y=1, xmin=0.39, xmax=0.62, c="blue", linestyle = "dashed");

font = {'family':'serif','color':'blue','size':16}
plt.title("Jimena García :)", loc = 'center', fontdict = font)

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
